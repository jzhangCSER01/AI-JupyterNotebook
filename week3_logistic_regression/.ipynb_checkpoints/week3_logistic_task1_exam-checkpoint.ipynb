# 逻辑回归预测考试通过
任务：
基于 examdata.csv 数据，建立逻辑回归模型 预测 Exam1 = 75, Exam2 = 60 时，该同学在 Exam3 是 passed or failed;
建立二阶边界，提高模型准确度

```python
# load the data
import pandas as pd
import numpy as np
data = pd.read_csv('examdata.csv')
data.head()
```

```python
# visualize the data
%matplotlib inline
from matplotlib import pyplot as plt
fig1 = plt.figure()
plt.scatter(data.loc[:, 'Exam1'], data.loc[:, 'Exam2'])
plt.title('Exam1-Exam2')
plt.xlabel('Exam1')
plt.ylabel('Exam2')
plt.show()
```

```python
# add label mask
mask = data.loc[:, 'Pass'] == 1
print(mask)
```

```python
fig2 = plt.figure()
passed = plt.scatter(data.loc[:, 'Exam1'][mask], data.loc[:, 'Exam2'][mask])
failed = plt.scatter(data.loc[:, 'Exam1'][~mask], data.loc[:, 'Exam2'][~mask])
plt.title('Exam1-Exam2')
plt.xlabel('Exam1')
plt.ylabel('Exam2')
plt.legend((passed, failed), ('passed', 'failed'))
plt.show()
```

```python
# define X y
X = data.drop(['Pass'], axis = 1)
y = data.loc[:, 'Pass']
X1 = data.loc[:, 'Exam1']
X2 = data.loc[:, 'Exam2']
X.head()
X1.head()
```

```python
print(X.shape, y.shape)
```

```python
# establish the model and train it
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression()
LR.fit(X, y)
```

```python
# show the predicted result and its accurary
y_predict = LR.predict(X)
print(y_predict)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y, y_predict)
print(accuracy)
```

```python
# exam1 = 70, exam2 = 65
y_test = LR.predict([[70, 65]])
print(y_test)
```

```python
theta0 = LR.intercept_
theta1, theta2 = LR.coef_[0][0], LR.coef_[0][1]
print(theta0, theta1, theta2)
```

```python
X2_new = -(theta0 + theta1 * X1) / theta2
print(X2_new)
```

```python
fig3 = plt.figure()
passed = plt.scatter(data.loc[:, 'Exam1'][mask], data.loc[:, 'Exam2'][mask])
failed = plt.scatter(data.loc[:, 'Exam1'][~mask], data.loc[:, 'Exam2'][~mask])
plt.plot(X1, X2_new)
plt.title('Exam1-Exam2')
plt.xlabel('Exam1')
plt.ylabel('Exam2')
plt.legend((passed, failed), ('passed', 'failed'))
plt.show()
```

```python
# create new data
X1_2 = X1 * X1
X2_2 = X2 * X2
X1X2 = X1 * X2
```

```python
X_new = {'X1' : X1, 'X2' : X2, 'X1_2' : X1_2, 'X2_2' : X2_2, 'X1X2' : X1X2}
X_new = pd.DataFrame(X_new)
print(X_new)
```

```python
# establish new model and train it
LR2 = LogisticRegression()
LR2.fit(X_new, y)
```

```python
y2_predict = LR2.predict(X_new)
accuracy2 = accuracy_score(y, y2_predict)
print(accuracy2)
```

```python
X1_new = X1.sort_values()
print(X1, X1_new)
```

```python
theta0 = LR2.intercept_
theta1, theta2, theta3, theta4, theta5 = LR2.coef_[0][0], LR2.coef_[0][1], LR2.coef_[0][2], LR2.coef_[0][3], LR2.coef_[0][4]
a = theta4
b = theta5 * X1_new + theta2
c = theta0 + theta1 * X1_new + theta3 * X1_new * X1_new
X2_new_boundary = (-b + np.sqrt(b * b - 4 * a * c)) / (2 * a)
#print(theta0, theta1, theta2, theta3, theta4, theta5)
print(X2_new_boundary)
```

```python
fig5 = plt.figure()
passed = plt.scatter(data.loc[:, 'Exam1'][mask], data.loc[:, 'Exam2'][mask])
failed = plt.scatter(data.loc[:, 'Exam1'][~mask], data.loc[:, 'Exam2'][~mask])
plt.plot(X1_new, X2_new_boundary)
plt.title('Exam1-Exam2')
plt.xlabel('Exam1')
plt.ylabel('Exam2')
plt.legend((passed, failed), ('passed', 'failed'))
plt.show()
```
